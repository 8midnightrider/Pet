# 4) Здесь проблема заключается в том, что source airport ID(destination airport ID) в routes является id в airports, но мы видим, что для одного и того же кода IATA в airports айди есть, а в routes его нет
# То есть они говорят, что аэропорт в базе числиться и не числиться одновременно
## Сильно.

```python
r1 = routes[routes['source airport ID'] == '\\N']
r2 = routes[routes['destination airport ID'] == '\\N']

r1 = r1[['source airport', 'source airport ID']]
r1 = r1.merge(airports[['IATA', 'id']], left_on = 'source airport', right_on = 'IATA')
r1 = r1[['source airport', 'IATA', 'id', 'source airport ID']]

r1
```

```python
r2 = r2[['destination airport', 'destination airport ID']]
r2 = r2.merge(airports[['IATA', 'id']], left_on = 'destination airport', right_on = 'IATA')
r2 = r2[['destination airport', 'IATA', 'id', 'destination airport ID']]
 
r2
```
