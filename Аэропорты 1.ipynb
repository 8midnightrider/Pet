```python
import pandas as pd
import numpy as np
```

```python
#1
airports = pd.read_csv('https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat', header = None)
routes = pd.read_csv('https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat', header = None)
```

```python
airports.columns = ['id', 'name', 'city', 'country', 'IATA', 'ICAO', 'latitude', 'longitude', 'altitude', 'timezone', 'DST',
                   'tz db', 'type', 'source']
routes.columns = ['airline', 'airline ID', 'source airport', 'source airport ID', 'destination airport',
                 'destination airport ID', 'codeshare', 'stops', 'equipment']
airports = airports.where(airports != '\\N', np.nan)
routes = routes.where(routes != np.nan)

airports
```

```python
routes
```

```python
routes1 = pd.DataFrame({'departures' : routes['source airport'].value_counts()})
routes2 = pd.DataFrame({'arrivals' : routes['destination airport'].value_counts()})
routes1['airport IATA'] = routes1.index
routes2['airport IATA'] = routes2.index

rout_res = routes1.merge(routes2, left_on = 'airport IATA', right_on = 'airport IATA')
```

```python
rout_res = rout_res[['airport IATA', 'departures', 'arrivals']]
rout_res['sum'] = rout_res['departures'] + rout_res['arrivals']
```

```python
rout_res = rout_res.merge(airports[['IATA', 'country', 'name']], left_on = 'airport IATA', right_on = 'IATA')
rout_res
```

```python
rout_res = rout_res.set_index(['country', 'name'])
```

```python
top10airports = rout_res.nlargest(10, 'sum')
top10airports                                    #10 крупнейших аэропортов по сумме вылет/прилет
```
