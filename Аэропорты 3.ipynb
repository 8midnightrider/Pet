```python
import pandas as pd 
import awoc
import folium as fl
```

```python
#3
airports = pd.read_csv('https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat', header = None)
routes = pd.read_csv('https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat', header = None)
```

```python
airports.columns = ['id', 'Airport_name', 'City', 'Country', 'IATA', 'ICAO', 'Широта', 'Долгота', 'altitude', 'timezone', 'DST',
                   'tz db', 'type', 'source']
routes.columns = ['Airline', 'airline ID', 'Вылет из', 'destination airport ID', 'Прилет в',
                 'destination airport ID', 'codeshare', 'stops', 'equipment']

airports = airports[airports['IATA'] != '\\N']
```

```python
airports = airports[['Airport_name', 'City', 'Country', 'IATA', 'Широта', 'Долгота']] 
routes = routes[['Airline', 'Вылет из','Прилет в']]
# остальные поля как будто и не нужны, а так глазу приятнее
```

```python
my_world = awoc.AWOC()
countr = my_world.get_countries()
countr = pd.DataFrame(countr)
countr = countr[['Country Name', 'Continent Name']]
airports = airports.merge(countr, left_on='Country', right_on='Country Name')
airports.drop(columns = 'Country Name')
airports = airports[['Continent Name', 'Country', 'Airport_name', 'IATA', 'Широта', 'Долгота']]
```

```python
airports
```

```python
eu_airports = airports[airports['Continent Name'] == 'Europe']
af_airports = airports[airports['Continent Name'] == 'Africa']
na_airports = airports[airports['Continent Name'] == 'North America']
sa_airports = airports[airports['Continent Name'] == 'South America']
as_airports = airports[airports['Continent Name'] == 'Asia']
oc_airports = airports[airports['Continent Name'] == 'Oceania']
an_airports = airports[airports['Continent Name'] == 'Antarctica']
```

```python
routes1 = pd.DataFrame({'departures' : routes['Вылет из'].value_counts()})
routes2 = pd.DataFrame({'arrivals' : routes['Прилет в'].value_counts()})
routes1['airport IATA'] = routes1.index
routes2['airport IATA'] = routes2.index

rout_res = routes1.merge(routes2, left_on = 'airport IATA', right_on = 'airport IATA')
```

```python
rout_res = rout_res.merge(airports[['IATA', 'Country', 'Airport_name']], left_on = 'airport IATA', right_on = 'IATA')
rout_res['sum'] = rout_res['departures'] + rout_res['arrivals']
```

```python
rout_res = rout_res[['IATA', 'sum']]
```

```python
eu_airports = eu_airports.merge(rout_res, left_on='IATA', right_on='IATA')
eu_airports
```
